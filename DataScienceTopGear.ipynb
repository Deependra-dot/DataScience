```python
%matplotlib inline
import pandas as pd
import numpy as np
from pylab import *
import matplotlib.pyplot as plt

r_cols = ['population', 'profit']
input_data=pd.read_csv('F:/PythonCoding/ex1data1.txt',sep=',',names=r_cols, usecols=range(2), encoding="ISO-8859-1")
#input_data.head()
population=input_data.population
profit=input_data.profit
scatter(population,profit)
plt.plot(population,profit)
plt.show()
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(population, profit)

def predict(x):
    return slope * x + intercept

fitLine = predict(population)

plt.scatter(population, profit)
plt.plot(population, fitLine, c='b')
plt.show()
```

```python
x = np.array(population)
y = np.array(profit)

p4 = np.poly1d(np.polyfit(x, y, 3))
xp = np.linspace(4, 25, 25)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```
